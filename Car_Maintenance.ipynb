```python
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/MohdRazidi/Car-Maintenance/refs/heads/main/Car%20Service%20Maintenance%20.csv"
df = pd.read_csv(url)
```

```python
df.head()
```

```python
print("Dataset shape:", df.shape)
```

```python
df.isnull().sum()
```

```python
print("\nMissing values per column:\n", df.isnull().sum())
```

```python
mileage_cols = ['1000KM','5000KM','10000KM','20000KM','30000KM','40000KM',
                '50000KM','60000KM','70000KM','80000KM','90000KM','100000KM']

df[mileage_cols] = df[mileage_cols].fillna(0)
```

```python
print("\nMissing values per column:\n", df.isnull().sum())
```

```python
# Display first 10 columns
df.iloc[:, :10].head(10)
```

```python
# Perodua data (already Peninsular & EM)
perodua_df = df[df['maker'] == 'Perodua']

# Other makers
others_df = df[df['maker'] != 'Perodua']
```

```python
# For others, duplicate Nationwide row into Peninsular and EM
others_peninsular = others_df.copy()
others_peninsular['region'] = 'Peninsular'

others_em = others_df.copy()
others_em['region'] = 'EM'

# Combine the duplicated rows
others_df_expanded = pd.concat([others_peninsular, others_em], ignore_index=True)
```

```python
# Combine with Perodua data
df_standardized = pd.concat([perodua_df, others_df_expanded], ignore_index=True)

# Optional: sort for readability
df_standardized.sort_values(by=['maker','model','variant','region'], inplace=True)
df_standardized.reset_index(drop=True, inplace=True)
```

```python
# Set option to display all rows
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)  # show all columns too

# Display all data
df
```

```python
df_standardized.to_csv("Car_Service_Maintenance_Standardized.csv", index=False)
```
